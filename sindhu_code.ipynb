```python
df = df.drop("Time", axis=1)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
```

```python
df['std_Amount'] = scaler.fit_transform(df['Amount'].values.reshape (-1,1))
```

```python
df = df.drop("Amount", axis=1)
```

```python
import imblearn
from imblearn.under_sampling import RandomUnderSampler
```

```python
undersample = RandomUnderSampler(sampling_strategy=0.5)
```

```python
cols = df.columns.tolist()
cols = [c for c in cols if c not in ["Class"]]
target = "Class"
```

```python
X = df[cols]
Y = df[target]
```

```python
X_under, Y_under = undersample.fit_resample(X, Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_under, Y_under, test_size=0.2, random_state=1)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix,roc_curve,roc_auc_score,auc,precision_recall_curve
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression = LogisticRegression(random_state=2)
```

```python
logistic_regression = logistic_regression.fit(X_train, y_train)
```

```python
y_pred_logistic_regression = logistic_regression.predict(X_test)
```

```python
y_pred_logistic_proba = logistic_regression.predict_proba(X_test)[::,1]
fpr_logistic, tpr_logistic, _ = metrics.roc_curve(y_test,  y_pred_logistic_proba)
auc_logistic = metrics.roc_auc_score(y_test, y_pred_logistic_proba)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_logistic_regression))
print("Precision:",metrics.precision_score(y_test, y_pred_logistic_regression))
print("Recall:",metrics.recall_score(y_test, y_pred_logistic_regression))
print("F1 Score:",metrics.f1_score(y_test, y_pred_logistic_regression))
print("AUC Logistic Regression :", auc_logistic)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.lineplot(x=fpr_logistic, y=tpr_logistic, label=f"Logistic Regression, AUC={auc_logistic:.3f}", ax=axes[0])
axes[0].plot([0, 1], [0, 1], 'k--')
axes[0].set_xlabel('False positive rate')
axes[0].set_ylabel('True positive rate')
axes[0].set_title('Logistic Regression ROC curve')
axes[0].legend(loc=4)
logit_precision, logit_recall, _ = precision_recall_curve(y_test, y_pred_logistic_proba)
no_skill = len(y_test[y_test == 1]) / len(y_test)
sns.lineplot(x=logit_recall, y=logit_precision, label='Logistic', ax=axes[1])
axes[1].plot([0, 1], [no_skill, no_skill], linestyle='--', color='red', label='No Skill')
axes[1].set_xlabel('Recall')
axes[1].set_ylabel('Precision')
axes[1].set_title('Precision-Recall curve')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from keras.layers import Dropout
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32, input_shape=(29,), activation='relu')),
model.add(Dropout(0.2)),
model.add(Dense(16, activation='relu')),
model.add(Dropout(0.2)),
model.add(Dense(8, activation='relu')),
model.add(Dropout(0.2)),
model.add(Dense(4, activation='relu')),
model.add(Dropout(0.2)),
model.add(Dense(1, activation='sigmoid'))
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
```

```python
model.compile(optimizer=opt, loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
earlystopper = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', min_delta=0, patience=15, verbose=1,mode='auto', baseline=None, restore_best_weights=False)
```

```python
history = model.fit(X_train.values, y_train.values, epochs = 50, batch_size=5, validation_split = 0.15, verbose = 0,
                    callbacks = [earlystopper])
history_dict = history.history
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.lineplot(data=history_dict, x=range(1, len(history_dict['loss']) + 1), y='loss', label='Training Loss', color='blue', alpha = 0.8)
sns.lineplot(data=history_dict, x=range(1, len(history_dict['val_loss']) + 1), y='val_loss', label='Validation Loss', color='red', alpha = 0.8)
plt.legend()
plt.xlabel("Epochs")
plt.title("Training and Validation Loss")
plt.subplot(1, 2, 2)
sns.lineplot(data=history_dict, x=range(1, len(history_dict['accuracy']) + 1), y='accuracy', label='Training Accuracy', color='blue', alpha = 0.8)
sns.lineplot(data=history_dict, x=range(1, len(history_dict['val_accuracy']) + 1), y='val_accuracy', label='Validation Accuracy', color='red', alpha = 0.8)
plt.legend()
plt.xlabel("Epochs")
plt.title("Training and Validation Accuracy")

plt.show()
```
